```python
#rename columns
pd_rbd = pd_rbd.rename(columns=columns)
```

```python
#remove columns with only 1 value
def remove_cols(df):
    remove = []
    for col in df:
        if len(df[col].unique()) == 1:
            remove.append(col)
    df = df.drop(columns=remove)
    return df
```

```python
pd_rbd = remove_cols(pd_rbd)
```

```python
#rename and encode id column
pd_rbd['id'] = pd_rbd['id'].str[:2]
pd_rbd = pd_rbd.rename(columns={'id':'status'})
```

```python
#encode categorical data
encode = {'status':{'PD':1, 'RB':2, 'HC':0},
          'gender':{'F':0, 'M':1},
          'fam_hist': {'No':0, 'Yes':1, '-':-1},
          'depress_med':{'No':0, 'Yes (Aurorix)':1, 'Yes (Trittico)':2, 'Yes (Remood)':3,
                         'Yes (Zoloft)':4, 'Yes (Anafranil)':5, 'Yes (Citalec)':6,
                         'Yes (Velaxin)':7, 'Yes (Asentra)':8, 'Yes (Cipralex)':9 },
          'benz_med':{'No':0, 'Yes (Rivotril)':1, 'Yes (Xanax)':2, 'Yes (Neurol)':3}
          }
pd_rbd.replace(encode, inplace=True)
```

```python
#fill in rows with now data with nan
pd_rbd = pd_rbd.replace('-', np.nan)
```

```python
#convert data types
pd_rbd['hy_scale'] = pd_rbd['hy_scale'].astype(float)
pd_rbd['age'] = pd_rbd['age'].astype(int)
```

```python
#create age range column
bins = [30, 40, 50, 60, 70, 80, 90]
labels = ['30-39', '40-49', '50-59', '60-69', '70-79','80+']
pd_rbd['age_range'] = pd.cut(pd_rbd.age, bins, labels = labels,include_lowest = True)
```

```python
#rearrange columns
cols = pd_rbd.columns.tolist()
cols = cols[:2] + cols[-1:] + cols[2:62]
pd_rbd_final = pd_rbd[cols]
```
